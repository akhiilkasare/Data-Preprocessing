## Feature Scaling

```python
#It is used to scale the data in a particular range for ex:-1 to 1
#why do we need this?Because if the data is not scaled not normalized we'll not be able to get the desired result or
#or the output can be miss leading 
```

```python
from sklearn import preprocessing
```

```python
x = np.array([[-400],[-100],[0],[100],[400]])
```

```python
x
```

```python
minmaxscaler = preprocessing.MinMaxScaler(feature_range=(0,1))
```

```python
x_scaler = minmaxscaler.fit_transform(x)
```

```python
x_scaler
```

```python
# Formula behind this
# (Xi -Xmin) / (Xmax - Xmin)
# (-100-(-400)) / (400-(-400))
# (-100 + 400) / (400 + 400)
# 300/800 = 3/8 = 0.375
```
